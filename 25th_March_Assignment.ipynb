## Q2.
### What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
import matplotlib.pyplot as plt
import pandas as pd



# Load the dataset into a DataFrame
flight_data = pd.read_excel('flight_price.xlsx')

# Extract the 'price' column (replace 'price' with the actual column name in your dataset)
prices = flight_data['Price']

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(prices, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

## Q3.
### What is the range of prices in the dataset? What is the minimum and maximum price?

```python
# Calculate the minimum and maximum prices
min_price = prices.min()
max_price = prices.max()

# Calculate the range
price_range = max_price - min_price

# Display the results
print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("Price Range:", price_range)
```

## Q5. 
### Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.



1. **Create a Boxplot:**
   - A boxplot visually represents the distribution of your data.
   - The box represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3).
   - Outliers can be identified as individual points beyond a certain distance (commonly 1.5 times the IQR) from the edges of the box.

2. **Identify Potential Outliers:**
   - Any data points outside the "whiskers" of the boxplot may be considered potential outliers.
   - Some boxplot implementations use different methods to determine the threshold for identifying outliers.

3. **Evaluate Impact on Analysis:**
   - Outliers can significantly impact statistical analyses, as they can skew means, standard deviations, and other summary statistics.
   - Consider whether outliers are valid data points or errors in data collection.
   - Decide whether to include or exclude outliers based on the goals of your analysis and the nature of the data.

4. **Sensitivity Analysis:**
   - Perform a sensitivity analysis by conducting your analysis both with and without outliers.
   - Assess the robustness of your results to determine whether outliers have a substantial impact on your conclusions.

Remember that the decision to treat a data point as an outlier depends on the specific context of your analysis and the characteristics of your dataset. It's essential to use your domain knowledge and judgment to determine the appropriate handling of outliers in your particular case.

**Visualization:** If you plan to create visualizations, missing values might affect the accuracy and interpretability of your visualizations.

**Model Performance:** If you're building predictive models, missing values can impact the performance of machine learning algorithms. Some algorithms handle missing data better than others, and imputation or removal strategies may be necessary.

**Insights and Recommendations:** Missing values may affect the accuracy of insights and recommendations derived from your analysis. It's essential to be transparent about the presence of missing data when presenting your findings.

### What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
# Filter out rows with missing values in 'Size' or 'Rating'
filtered_df = playstore_df.dropna(subset=['Size', 'Rating'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Size'], filtered_df['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('App Size')
plt.ylabel('App Rating')
plt.show()
```

## Q13.
### How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Load the dataset into a DataFrame
playstore_df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Group the data by 'Type' and calculate the average price for each type
average_prices = playstore_df.groupby('Type')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(8, 5))
average_prices.plot(kind='bar', color=['blue', 'green'], alpha=0.7)
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Wine dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
column_names = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols",
                "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue",
                "OD280/OD315 of diluted wines", "Proline"]

df = pd.read_csv(url, header=None, names=column_names)

# Split the dataset into features and target variables
X = df.drop("Class", axis=1)
y = df["Class"]

# Perform data preprocessing (scaling)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA on the preprocessed dataset
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Determine the optimal number of principal components
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Plot the explained variance ratio
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio')
plt.show()

# Choose the optimal number of principal components based on the plot
# You can manually inspect the plot and define optimal_number_of_components

# Visualize the results of PCA using a scatter plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Results - Scatter Plot')
plt.show()

# Define the optimal number of clusters based on your analysis
optimal_number_of_clusters = 3  # Replace with your chosen number of clusters

# Perform clustering on the PCA-transformed data using K-Means clustering algorithm
kmeans = KMeans(n_clusters=optimal_number_of_clusters)
clusters = kmeans.fit_predict(X_pca)

# Visualize the clusters on the PCA scatter plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Results with Clusters - Scatter Plot')
plt.show()

# Interpret the results of PCA and clustering analysis
# Analyze the clusters and their characteristics based on the original target variable or other criteria
```

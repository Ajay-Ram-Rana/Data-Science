# Q1. 
## Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

````python
import scipy.stats as stats

# Sample mean and standard deviation
sample_mean = 50
sample_std_dev = 5

# Sample size (you need to specify this)
sample_size = 30

# Confidence level
confidence_level = 0.95

# Calculate the standard error (standard deviation of the sample mean)
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1) * standard_error

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print Margin of error 
print("with margin of error : {:.3f}".format(margin_of_error))
# Print the results
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
````

***Interpreting the results:***
The 95% confidence interval for this sample is given as (lower_bound, upper_bound). In this case, it might look something like this:

"Based on the sample data, we can say with 95% confidence that the population mean lies between lower_bound and upper_bound."

For example, if the calculated interval is (47.52, 52.48), it means that you are 95% confident that the true population mean falls within this range. In other words, if you were to take many samples and calculate 95% confidence intervals for each, approximately 95% of them would contain the true population mean.

# Q2. 
## Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

To conduct a chi-square goodness of fit test in Python to determine if the distribution of colors of M&Ms in a bag matches the expected distribution, you can use the `scipy.stats` library. Here's a step-by-step guide on how to perform this test:

1. Define your observed and expected distributions.
2. Calculate the chi-square statistic.
3. Determine the degrees of freedom.
4. Find the p-value.
5. Compare the p-value to the significance level.

Here's the Python code to perform the test:

```python
import numpy as np
from scipy.stats import chi2_contingency

# Define observed counts of M&M colors
observed_counts = np.array([18, 22, 15, 9, 11, 25])  # Replace with your actual counts

# Define the expected distribution
expected_distribution = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

# Calculate the expected counts based on the expected distribution
total_observed = np.sum(observed_counts)
expected_counts = expected_distribution * total_observed

# Perform the chi-square goodness of fit test
chi2, p, dof, expected = chi2_contingency([observed_counts])

# Set the significance level
alpha = 0.05

# Print the results
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")

# Check for statistical significance
if p < alpha:
    print("Reject the null hypothesis: The distributions are not the same.")
else:
    print("Fail to reject the null hypothesis: The distributions match the expected distribution.")
```

Make sure to replace `observed_counts` with the actual counts you have in your M&M bag. The code above calculates the chi-square statistic, degrees of freedom, and p-value and compares it to the specified significance level (0.05). If the p-value is less than the significance level, you reject the null hypothesis, indicating that the observed distribution does not match the expected distribution. Otherwise, you fail to reject the null hypothesis, suggesting that they match.

````python
import numpy as np
from scipy.stats import chi2_contingency

observed_counts = np.array([18, 22, 15, 9, 11, 25])
expected_distribution = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

total_observed = np.sum(observed_counts)
expected_counts = expected_distribution * total_observed

chi2, p, dof, expected = chi2_contingency([observed_counts])


alpha = 0.05

# Print the results
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")

# Check for statistical significance
if p < alpha:
    print("Reject the null hypothesis: The distributions are not the same.")
else:
    print("Fail to reject the null hypothesis: The distributions match the expected distribution.")
````

# Q4. 
### A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

To calculate a confidence interval for the true proportion of individuals in the population who smoke, you can use the binomial proportion confidence interval formula. In Python, you can do this with the scipy.stats library. Here's how you can calculate the 95% confidence interval:

````python
import scipy.stats as stats

# Sample size
n = 500

# Number of individuals who smoke
x = 60

# Calculate the sample proportion
p_hat = x / n

# Set the confidence level and calculate the standard error
confidence_level = 0.95
alpha = 1 - confidence_level
z = stats.norm.ppf(1 - alpha / 2)
se = np.sqrt((p_hat * (1 - p_hat)) / n)

# Calculate the margin of error
margin_of_error = z * se

# Calculate the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

# Print the confidence interval
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
````

This code calculates the 95% confidence interval for the true proportion of individuals in the population who smoke. Make sure to change the values of n and x to match your specific data. The norm.ppf function is used to find the Z-score for a given confidence level.

The confidence interval is presented in the form (lower_bound, upper_bound), where the true proportion of smokers is expected to fall within that range with 95% confidence.

## Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

To calculate a 90% confidence interval for a sample with a mean of 75 and a standard deviation of 12, you can use the scipy.stats library in Python. The confidence interval is calculated using the t-distribution because the population standard deviation is not known. Here's how you can calculate the confidence interval:

````python
import scipy.stats as stats

# Sample mean
sample_mean = 75

# Sample standard deviation
sample_std = 12

# Sample size (you may need to specify this)
sample_size =  120 # Insert your sample size

# Confidence level
confidence_level = 0.90

# Degrees of freedom (for a sample, df = sample_size - 1)
df = sample_size - 1

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df) * (sample_std / (sample_size**0.5))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the confidence interval
print(f"90% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
````

This code calculates the 90% confidence interval for the sample mean. You need to specify the sample_size with your actual sample size. The t.ppf function is used to find the critical t-value for the given confidence level and degrees of freedom.

***Interpretation:***
The 90% confidence interval for the sample mean (assuming the population follows a normal distribution) is a range of values from the lower_bound to the upper_bound. It means that you can be 90% confident that the true population mean falls within this interval. In your case, the true population mean is expected to be between the lower_bound and the upper_bound with 90% confidence.

# Q6. 
### Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

You can use Python and the scipy.stats library to plot the chi-square distribution with 10 degrees of freedom and shade the area corresponding to a chi-square statistic of 15. Here's how you can do it:

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Generate values for the chi-square distribution
x = np.linspace(0, 30, 1000)  # Adjust the range as needed

# Calculate the chi-square probability density function (PDF)
pdf = stats.chi2.pdf(x, df)

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Plot the chi-square distribution
ax.plot(x, pdf, label=f'Chi-Square ({df} df)')

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(0, 15, 1000)
shade = stats.chi2.pdf(x_shade, df)
ax.fill_between(x_shade, shade, color='gray', alpha=0.5, label='Chi-Square < 15')

# Label the axes and add a legend
ax.set_xlabel('Chi-Square Statistic')
ax.set_ylabel('Probability Density')
ax.set_title(f'Chi-Square Distribution ({df} df)')
ax.legend()

# Display the plot
plt.show()
````

In this code, we first generate values for the chi-square distribution with 10 degrees of freedom. We then calculate the probability density function (PDF) for these values. The code also shades the area under the curve for chi-square values less than 15.

Adjust the range and other parameters as needed for your specific requirements.

## A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

To calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke based on a sample of 1000 people, you can use the binomial proportion confidence interval formula. Here's how you can do it in Python:

````python
import scipy.stats as stats

# Sample size
n = 1000

# Number of people who preferred Coke in the sample
x = 520

# Calculate the sample proportion
p_hat = x / n

# Set the confidence level and calculate the standard error
confidence_level = 0.99
alpha = 1 - confidence_level
z = stats.norm.ppf(1 - alpha / 2)
se = (p_hat * (1 - p_hat) / n)**0.5

# Calculate the margin of error
margin_of_error = z * se

# Calculate the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

# Print the confidence interval
print(f"99% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
````

This code calculates the 99% confidence interval for the true proportion of people in the population who prefer Coke based on your sample data. It uses the normal distribution approximation since the sample size is large. The confidence interval is presented in the form (lower_bound, upper_bound), where you can be 99% confident that the true proportion of people in the population who prefer Coke falls within this interval.

Make sure to replace the values of n and x with your specific sample size and the number of people who preferred Coke in your sample.

### A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

To conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin, you can use Python. The null hypothesis (H0) in this case is that the coin is fair, while the alternative hypothesis (H1) is that the coin is biased towards tails. You'll compare the observed frequencies with the expected frequencies for a fair coin.

Here's how you can perform the chi-square test in Python:

````python
import scipy.stats as stats

# Observed frequencies
observed_tails = 45
observed_heads = 100 - observed_tails

# Expected frequencies for a fair coin
expected_tails = 0.5 * 100
expected_heads = 100 - expected_tails

# Create an observed-expected frequency table
observed = [observed_tails, observed_heads]
expected = [expected_tails, expected_heads]

# Perform the chi-square goodness of fit test
chi2, p = stats.chisquare(observed, expected)

# Set the significance level
alpha = 0.05

# Print the results
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")

# Check for statistical significance
if p < alpha:
    print("Reject the null hypothesis: The coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis: There is no evidence that the coin is biased towards tails.")
````

In this code, we calculate the observed and expected frequencies for tails and heads. Then, we perform the chi-square goodness of fit test using scipy.stats.chisquare. The null hypothesis is that the coin is fair, and we compare the p-value to the significance level (0.05). If the p-value is less than 0.05, you reject the null hypothesis, indicating that there is evidence that the coin is biased towards tails. Otherwise, you fail to reject the null hypothesis, suggesting that there is no evidence of bias.
